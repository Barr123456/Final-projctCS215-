# Ploting the legislations


```python
# import the neccesary library for the ploting.
# blow you see I imported matplotlib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

The code below first groups the data in the df_Guinea DataFrame by political party and calculates the total number of seats won by each party using the groupby() method and the sum() function. and the result is sorted in descending order using the sort_values() function with the ascending=False parameter. Then below I impliment a bar plot using Seaborn's sns.barplot() function to visualize seats won by each political party. The plot's size is set using the plt.figure() function, and the x-axis labels are rotated by 90 degrees using plt.xticks(). The plot's title and axis labels are set using plt.title(), plt.xlabel(), and plt.ylabel(). Finally, the plot is displayed using the plt.show() function.

```python
# Group the data by political party and calculate the number of seats won
seats_by_party = df_Guinea.groupby("party")["ma_seats"].sum().sort_values(ascending=False)
print(seats_by_party)

# Visualize the distribution of seats won by party using a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=seats_by_party.index, y=seats_by_party.values)
plt.xticks(rotation=90)
plt.title("Seats won by party")
plt.xlabel("Parties")
plt.ylabel("Number of seats")
plt.show()
```

```python
df_USA['date'] = pd.to_datetime(df_USA['date'])
df_USA['month'] = df_USA['date'].dt.month_name()
monthly_counts = df_USA.groupby('month')['killed'].sum()

# create a new column for the month
df_USA['month'] = df_USA['date'].dt.month_name()

# group the data by month and count the number of students killed
monthly_counts = df_USA.groupby('month')['killed'].sum()

# create a list of the month names in the correct order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# reindex the monthly_counts Series with the month_order list
monthly_counts = monthly_counts.reindex(month_order)

# print the monthly counts
print(monthly_counts)
```
